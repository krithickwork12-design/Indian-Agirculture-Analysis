Sorghum Production (Kharif and Rabi) by Region

```python
sorghum_region = (
    agri_data.groupby("state_name")[[
        "kharif_sorghum_production_(1000_tons)",
        "rabi_sorghum_production_(1000_tons)"
    ]].sum().reset_index()
)
top_sorghum_states = (
    sorghum_region
    .sort_values(
        by=["kharif_sorghum_production_(1000_tons)", "rabi_sorghum_production_(1000_tons)"],
        ascending=False
    ).head(10)
)

plt.figure(figsize=(10, 6))

x = np.arange(len(top_sorghum_states))
width = 0.35

plt.bar(
    x - width/2,top_sorghum_states["kharif_sorghum_production_(1000_tons)"],width,label="Kharif"
)

plt.bar(
    x + width/2,top_sorghum_states["rabi_sorghum_production_(1000_tons)"],width,label="Rabi"
)

plt.xticks(
    x,top_sorghum_states["state_name"],rotation=45
)

plt.xlabel("Region (State)")
plt.ylabel("Sorghum Production (1000 tons)")
plt.title("Sorghum Production (Kharif vs Rabi) by Region")
plt.legend()
plt.tight_layout()
plt.show()
```

OBSERVATION:
*   Maharastra has the highest sorghum production in both Kharif and Rabi across India.

Soybean Production by Top 5 States and Yield Efficiency

# AgriData Explorer – Final EDA Report
### Understanding Indian Agriculture using ICRISAT District-Level Data

This notebook presents the final Exploratory Data Analysis (EDA) findings derived from
India’s district-level agricultural dataset. The analysis focuses on crop production,
yield efficiency, regional dominance, long-term trends, and the relationship between
cultivated area and output.


## 1. Rice Production Analysis

- West Bengal produces the highest amount of rice in India.
- Rice production has remained higher than wheat production over the last 50 years.
- Within West Bengal, Medinipur district produces the highest quantity of rice among all districts.

### Interpretation
West Bengal’s dominance in rice production can be attributed to favorable climatic
conditions, fertile alluvial soil, and widespread irrigation facilities. District-level
analysis highlights production concentration within specific regions.


## 2. Wheat Production Analysis

- Uttar Pradesh is the largest wheat-producing state in India.
- Wheat production peaked in the year 2017, making it the highest wheat-producing year.
- West Bengal contributes approximately 38.5% of India’s total wheat production.

### Interpretation
Uttar Pradesh benefits from extensive canal irrigation and adoption of high-yield wheat
varieties, making it the national leader in wheat production.


## 3. Oilseed Production Insights

- Madhya Pradesh produces the highest amount of oilseeds in India.
- Madhya Pradesh also shows the highest yield efficiency among all states.
- This dominance is consistent across both total production and productivity metrics.

### Interpretation
Madhya Pradesh’s oilseed leadership highlights efficient land use, improved seed varieties,
and mechanized farming practices.


## 4. Sunflower and Groundnut Production

- Karnataka produces the highest amount of sunflower in India.
- Gujarat is the largest producer of groundnut across the country.

### Interpretation
Crop specialization across states is driven by soil type, rainfall patterns,
and regional agricultural practices.


## 5. Sugarcane Production – Long-Term Trend

- Sugarcane production has increased significantly over the last 50 years.
- Production has grown from approximately 10,000 tons to nearly 40,000 tons annually.

### Interpretation
The steady increase in sugarcane production reflects improved irrigation,
high-yield varieties, and rising industrial demand.


## 6. Millet Production Trends

- Pearl millet consistently shows higher production than finger millet.
- Pearl millet production has increased significantly over the last 50 years.
- Finger millet production has increased only marginally during the same period.

### Interpretation
Pearl millet’s drought tolerance and adaptability make it more scalable
than finger millet in semi-arid regions.


## 7. Sorghum Production (Kharif and Rabi)

- Maharashtra has the highest sorghum production in both Kharif and Rabi seasons across India.

### Interpretation
Sorghum remains a staple crop in semi-arid regions, especially where irrigation
availability is limited.


## 8. Area Cultivated vs Production Relationship

### Key Findings
- There is a strong positive correlation between cultivated area and production.
- Rice shows the strongest dependency on cultivated area.
- Wheat demonstrates moderate efficiency due to better irrigation.
- Maize shows higher yield variability.
- Area alone is not sufficient; yield efficiency and farming practices matter.

### Interpretation
Expanding cultivated land increases output, but sustainable growth depends
on improved yield and agricultural practices.


## 9. Yield Efficiency Comparison

- Wheat produces more yield per hectare compared to rice across all states.
- Uttar Pradesh records the highest wheat yield per hectare in India.

### Interpretation
Wheat benefits more from technological interventions such as improved seeds
and controlled irrigation, resulting in higher per-hectare productivity.


## 10. Key Takeaways

- Rice and wheat remain the backbone of Indian agriculture.
- Crop dominance varies significantly by region.
- Yield efficiency is a better performance indicator than total production.
- Millets and sorghum show strong potential for climate-resilient agriculture.
- Long-term trends reveal consistent growth in sugarcane and pearl millet production.


## 11. Business and Policy Implications

### For Farmers
- Focus on yield improvement rather than land expansion.
- Choose crops aligned with regional strengths.

### For Policymakers
- Target low-efficiency, high-area regions for intervention.
- Promote millets and sorghum for sustainable agriculture.
- Invest in irrigation and modern farming technologies.

### For Researchers
- Study climate impact on yield variability.
- Analyze efficiency gaps at district and regional levels.
