```python
#QUERY 1 -- Top 7 RICE PRODUCTION State Data(Bar_plot)
```

```python
state_rice_production = new_agri_data.groupby('State Name')['RICE PRODUCTION (1000 tons)'].sum().reset_index()
top_7_rice_states = state_rice_production.sort_values(by=rice_production,ascending=False).head(7)
plt.figure(figsize=(10, 7))
sns.barplot(
    x='RICE PRODUCTION (1000 tons)',
    y='State Name',
    data=top_7_rice_states,
    palette='Reds_d'
)
plt.title('Top 7 States by Total Rice Production (1966-2017)', fontsize=16)
plt.xlabel('Total Rice Production (1000 tons)', fontsize=12)
plt.ylabel('State Name', fontsize=12)
```

```python
#query 2 -- Top 5 Wheat Producing States Data(Bar_chart)and its percentage(%)(Pie_chart)
```

```python
state_wheat_production = new_agri_data.groupby('State Name')['WHEAT PRODUCTION (1000 tons)'].sum().reset_index()
top_5_wheat_states = state_wheat_production.sort_values(by='WHEAT PRODUCTION (1000 tons)',ascending=False).head(5)
total_national_production = state_wheat_production['WHEAT PRODUCTION (1000 tons)'].sum()
top_5_production_sum = top_5_wheat_states['WHEAT PRODUCTION (1000 tons)'].sum()
rest_of_india_production = total_national_production - top_5_production_sum

pie_data = top_5_wheat_states[['State Name', 'WHEAT PRODUCTION (1000 tons)']].copy()
rest_of_india_row = pd.DataFrame([['Rest of India', rest_of_india_production]],columns=['State Name', 'WHEAT PRODUCTION (1000 tons)'])
pie_data = pd.concat([pie_data, rest_of_india_row], ignore_index=True)
pie_data['Percentage'] = (pie_data['WHEAT PRODUCTION (1000 tons)'] / total_national_production) * 100


plt.figure(figsize=(10, 6))
sns.barplot(
    x='WHEAT PRODUCTION (1000 tons)',
    y='State Name',
    data=top_5_wheat_states,
    palette='Oranges_d'
)
plt.title('Top 5 States by Total Wheat Production (1966-2017)', fontsize=14)
plt.xlabel('Total Wheat Production (1000 tons)', fontsize=12)
plt.ylabel('State Name', fontsize=12)


plt.figure(figsize=(9, 9))
plt.pie(
    pie_data['WHEAT PRODUCTION (1000 tons)'],
    labels=pie_data['State Name'],
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette('pastel')
)
plt.title('Percentage Contribution of Top 5 States to Total National Wheat Production (1966-2017)', fontsize=14)
```

```python
#query 10 -- Sorghum Production (Kharif and Rabi) by Region
```

```python
KHARIF_PROD_COL = 'KHARIF SORGHUM PRODUCTION (1000 tons)'
RABI_PROD_COL = 'RABI SORGHUM PRODUCTION (1000 tons)'
STATE_COL = 'State Name'
state_seasonal_production = new_agri_data.groupby(STATE_COL).agg({KHARIF_PROD_COL: 'sum',RABI_PROD_COL: 'sum'}).reset_index()
state_seasonal_production['TOTAL_SORGHUM'] = state_seasonal_production[KHARIF_PROD_COL] + state_seasonal_production[RABI_PROD_COL]
state_seasonal_production = state_seasonal_production.sort_values(by='TOTAL_SORGHUM', ascending=False).drop(columns=['TOTAL_SORGHUM'])

plot_data = state_seasonal_production.melt(
    id_vars=[STATE_COL],
    value_vars=[KHARIF_PROD_COL, RABI_PROD_COL],
    var_name='Season',
    value_name='Total Production (1000 tons)'
)

plot_data['Season'] = plot_data['Season'].replace({KHARIF_PROD_COL: 'Kharif Sorghum',RABI_PROD_COL: 'Rabi Sorghum'})

plt.figure(figsize=(16, 8))
sns.barplot(
    x=STATE_COL,
    y='Total Production (1000 tons)',
    hue='Season',
    data=plot_data,
    palette={'Kharif Sorghum': 'orange', 'Rabi Sorghum': 'brown'}
)

plt.title('Kharif vs Rabi Sorghum Production by All States (1966-2017)', fontsize=16)
plt.xlabel('State Name', fontsize=12)
plt.ylabel('Total Production (1000 tons)', fontsize=12)
plt.xticks(rotation=70, ha='right') # Rotate labels for readability
plt.legend(title='Sorghum Season')
```

```python
#QUERY 14 - Impact of Area Cultivated on Production (Rice, Wheat, Maize)
```

```python

# Column names
RICE_PROD_COL = 'RICE PRODUCTION (1000 tons)'
RICE_AREA_COL = 'RICE AREA (1000 ha)'
WHEAT_PROD_COL = 'WHEAT PRODUCTION (1000 tons)'
WHEAT_AREA_COL = 'WHEAT AREA (1000 ha)'
MAIZE_PROD_COL = 'MAIZE PRODUCTION (1000 tons)'
MAIZE_AREA_COL = 'MAIZE AREA (1000 ha)'
YEAR_COL = 'Year'

# 1. Aggregate data to National Level for each year
yearly_data = new_agri_data.groupby(YEAR_COL).agg({
    RICE_AREA_COL: 'sum',
    RICE_PROD_COL: 'sum',
    WHEAT_AREA_COL: 'sum',
    WHEAT_PROD_COL: 'sum',
    MAIZE_AREA_COL: 'sum',
    MAIZE_PROD_COL: 'sum'
}).reset_index()

# 2. Determine the time frame
min_year = yearly_data[YEAR_COL].min()
max_year = yearly_data[YEAR_COL].max()

# --- Visualization: 3 Subplots for Area vs. Production ---
# Using 3 separate plots is necessary because the scales for each crop are vastly different
fig, axes = plt.subplots(3, 1, figsize=(10, 18))
fig.suptitle(f'Impact of Area on Production ({min_year}-{max_year})', fontsize=18, y=1.02)

# Plot 1: Rice
sns.regplot(
    ax=axes[0],
    data=yearly_data,
    x=RICE_AREA_COL,
    y=RICE_PROD_COL,
    color='blue',
    line_kws={'color': 'red'} # Highlight regression line
)
axes[0].set_title('Rice: Area vs. Production', fontsize=14)
axes[0].set_xlabel('Total Area (1000 ha)', fontsize=12)
axes[0].set_ylabel('Total Production (1000 tons)', fontsize=12)
axes[0].grid(True, linestyle='--', alpha=0.6)

# Plot 2: Wheat
sns.regplot(
    ax=axes[1],
    data=yearly_data,
    x=WHEAT_AREA_COL,
    y=WHEAT_PROD_COL,
    color='orange',
    line_kws={'color': 'red'}
)
axes[1].set_title('Wheat: Area vs. Production', fontsize=14)
axes[1].set_xlabel('Total Area (1000 ha)', fontsize=12)
axes[1].set_ylabel('Total Production (1000 tons)', fontsize=12)
axes[1].grid(True, linestyle='--', alpha=0.6)

# Plot 3: Maize
sns.regplot(
    ax=axes[2],
    data=yearly_data,
    x=MAIZE_AREA_COL,
    y=MAIZE_PROD_COL,
    color='purple',
    line_kws={'color': 'red'}
)
axes[2].set_title('Maize: Area vs. Production', fontsize=14)
axes[2].set_xlabel('Total Area (1000 ha)', fontsize=12)
axes[2].set_ylabel('Total Production (1000 tons)', fontsize=12)
axes[2].grid(True, linestyle='--', alpha=0.6)


# Print the data used for the charts (optional, but good practice)
print(f"National Aggregated Data ({min_year} to {max_year}):")
print(yearly_data.head().to_markdown(index=False, numalign="left", stralign="left"))
```
